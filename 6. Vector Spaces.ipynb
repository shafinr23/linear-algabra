## 1. Solutions To Linear Systems

We have 3 points in 2d space, and we would like to determine a second degree polynomial that passes through all of these points, having the form of:
$p(\chi) = \gamma_0 + \gamma_1\chi + \gamma_2\chi^2$. Solving this polynomial as a system of linear equations will result in either:

- A Solution
- No Solution
- Multiple Solutions

```python
points = np.array([(-2, -1), (0, 2), (2, 3)])
```

```python
plt.scatter(*zip(*points), color='b')
plt.xlim(-2.5, 2.5)
plt.ylim(-1.5, 3.5)
plt.grid()
plt.show()
```

### 1.1 A Solution

We know that:
    
- $p(-2) = -1$
- $p(0) = 2$
- $p(2) = 3$

So we can set up these equations:

$
\begin{array}{c}
p(-2) & = & \gamma_0 & + & \gamma_1(-2) & + & \gamma_2(-2)^2 & = & -1 \\
p(0) & = & \gamma_0 & + & \gamma_1(0) & + & \gamma_2(0)^2 & = & 2 \\
p(2) & = & \gamma_0 & + & \gamma_1(2) & + & \gamma_2(2)^2 & = & 3 \\
\end{array}
$

This linear system can be defined as:

$
\begin{bmatrix}
1 & -2 & 4 \\
1 & 0 & 0 \\
1 & 2 & 4 \\
\end{bmatrix}
$
$
\begin{bmatrix}
\gamma_0 \\
\gamma_1 \\
\gamma_2 \\
\end{bmatrix}
$
$ =
\begin{bmatrix}
-1 \\
2 \\
3 \\
\end{bmatrix}
$

Solving this system with elimination provides consistent equations, yielding:

$
\begin{bmatrix}
\gamma_0 \\
\gamma_1 \\
\gamma_2
\end{bmatrix}
$
$ =
\begin{bmatrix}
2 \\
1 \\
-\frac{1}{4}
\end{bmatrix}
$

So that:

$p(\chi) = 2 + \chi - \frac{1}{4}\chi^2$

You can capture the polynomials at the values $-2, 0,$ and $2$:

```python
def p(x):
    return 2 + x - 0.25 * x**2

def gamma0(x):
    return 1
def gamma1(x):
    return x
def gamma2(x):
    return x**2

x = np.linspace(points[0][0], points[-1][0])
y = list(map(p, x))
g0 = list(map(gamma0, x))
g1 = list(map(gamma1, x))
g2 = list(map(gamma2, x))

plt.scatter(*zip(*points), color='b')
plt.plot(x, y, color='b')
plt.plot(x, g0, color='g')
plt.plot(x, g1, color='y')
plt.plot(x, g2, color='r')

plt.xlim(-2.5, 2.5)
plt.ylim(-1.5, 3.5)
plt.grid()
plt.show()
```

But what happens if a free variable is encountered? And what happens if elimination provides inconistent equations?

### 1.2 No Solution

Let's add an extra point to the set $(4, 9)$, so that we have four points in 2d space:

```python
points = np.array([(-2, -1), (0, 2), (2, 3), (4, 9)])
```

We would have known that:
    
- $p(-2) = -1$
- $p(0) = 2$
- $p(2) = 3$
- $p(4) = 9$

Meaning that our polynomial would have to go through all of these points. And we would end up with four equations and three unknowns:

$
\begin{bmatrix}
1 & -2 & 4 \\
1 & 0 & 0 \\
1 & 2 & 4 \\
1 & 4 & 16 \\
\end{bmatrix}
$
$
\begin{bmatrix}
\gamma_0 \\
\gamma_1 \\
\gamma_2
\end{bmatrix}
$
$ =
\begin{bmatrix}
-1 \\
2 \\
3 \\
9 \\
\end{bmatrix}
$

We have already found a solution for the unknowns, so we can plug it in and see if 9 pops out:

$
\begin{bmatrix}
1 & 4 & 16 \\
\end{bmatrix}
$
$
\begin{bmatrix}
2 \\
1 \\
-\frac{1}{4}
\end{bmatrix}
$
$ \neq
\begin{bmatrix}
9 \\
\end{bmatrix}
$

As we can see, this results in an inconsistent equation. $Ax = b$ has a solution if and only if $b$ is in the **column space** of $A$. This will be explained after we dealt with the multiple solutions case.

### 1.3 Multiple Solutions

Take the following system:

$
\begin{bmatrix}
2 & 2 & -2 \\
-2 & -3 & 4 \\
4 & 3 & -2 \\
\end{bmatrix}
\begin{bmatrix}
\chi_0 \\
\chi_1 \\
\chi_2
\end{bmatrix}
$
$ =
\begin{bmatrix}
0 \\
3 \\
3
\end{bmatrix}
$

First we apply gauss-jordan elimination:

$
\left[
\begin{array}{ccc|c}
2 & 2 & -2 & 0 \\
-2 & -3 & 4 & 3\\
4 & 3 & -2 & 3 \\
\end{array}
\right]
$
$ \rightarrow
\left[
\begin{array}{ccc|c}
2 & 2 & -2 & 0 \\
0 & -1 & 2 & 3 \\
0 & -1 & 2 & 3 \\
\end{array}
\right]
$
$ \rightarrow
\left[
\begin{array}{ccc|c}
2 & 0 & 2 & 6 \\
0 & -1 & 2 & 3 \\
0 & 0 & 0 & 0 \\
\end{array}
\right]
$

Notice that we end up with two equations and three unkowns, meaning that it is under-determined. We work ahead until we have ones on the diagonal:

$ \rightarrow
\left[
\begin{array}{ccc|c}
1 & 0 & 1 & 3 \\
0 & 1 & -2 & -3 \\
0 & 0 & 0 & 0 \\
\end{array}
\right]
$

And we translate it back into a linear system:

$
\begin{array}{c}
\chi_0 & & + & \chi_2 & = & 3 \\
& \chi_1 & - & 2\chi_2 & = & -3 \\
& & & 0 & = & 0
\end{array}
$

We can see that there are no inconsistencies, $0 = 0$, is consistent. We call $\chi_2$ a free variable, and the other variables dependent variables. The free variable can take on any value, so we will call any value $\beta$:

$
\begin{array}{c}
\chi_0 & & + & \beta & = & 3 \\
& \chi_1 & - & 2\beta & = & -3 \\
& & & \chi_2 & = & \beta
\end{array}
$

Now we can continue solving like we normally would:

$ \rightarrow
\left[
\begin{array}{ccc|cc}
1 & 0 & 0 & 3-\beta \\
0 & 1 & 0 & -3+2\beta\\
0 & 0 & 1 & 0+\beta \\
\end{array}
\right]
$

We can write this solution as:

$
\begin{bmatrix}
\chi_0 \\
\chi_1 \\
\chi_2
\end{bmatrix}
$
$ =
\begin{bmatrix}
3 \\
-3 \\
0
\end{bmatrix}
$
$ + \beta
\begin{bmatrix}
-1 \\
2 \\
1
\end{bmatrix}
$

We know that $\beta$ can take up any value, including zero. Therefore we know that the solution is:

$
\begin{bmatrix}
3 \\
-3 \\
0
\end{bmatrix}
$
and any number of this 
$
\begin{bmatrix}
-1 \\
2 \\
1
\end{bmatrix}
$
vector added or subtracted from it. So we know this particular system has an infinite amount of solutions.

## 2. Vector Space

A vector space is a set of vectors. For example, $\mathbb{R}^n$ is a $n$-dimensional vector space. All vectors in that dimension are in $\mathbb{R}^n$. Any subset $S$ of $\mathbb{R}^n$ is also a vector space, having the following properties:

- The zero vector must be in $S$
- If $v, w \in S$ then $(v+w)\in S$
- If $\alpha \in \mathbb{R}$ and $v \in S$ then $\alpha v \in S$

### 2.1 Subspace

A subspace is a subset of a vector space that is itself a vector space. Let's have a look at three examples:

<br>
<font color="red">
$$\mathbb{R}^2 = \{x \mid x = 
\begin{pmatrix}
\chi_0 \\
\chi_1 
\end{pmatrix}
\}$$
</font>
<br>
<font>
$$S = \{x \mid x = 
\begin{pmatrix}
\chi_0 \\
0 \\
\end{pmatrix}
\}$$
</font>
<br>
<font color="blue">
$$T = \{x \mid x = 
\begin{pmatrix}
\chi_0 \\
1 \\
\end{pmatrix}
\}$$
</font>

$\mathbb{R}^2$ is itself a subspace of $\mathbb{R}^2$. $S$ is a subspace, because it is a vector space, and it's a subset of $\mathbb{R}^n$. $T$ is **NOT** a subspace, because the zero vector is not in it. And because adding or scaling a vector results in a vector outside $T$. Take a look at the blue arrow that's scaled in the following plot:

```python
import random
random.seed(555)
r2 = [[random.uniform(-25, 25), random.uniform(-2, 2)] for x in range(10)]
s = [[x, 0] for x in np.linspace(-20, 20, 8)]
t = [[x, 1] for x in np.linspace(-25, 25, 8)]
t.append([x * 1.8 for x in t[3]]) # Scaling a vector in T, showing it's not a subspace

fig, ax = plt.subplots(1)
ax.set_facecolor('xkcd:salmon')
plt.quiver([0], [0], *zip(*s), angles = 'xy', scale_units = 'xy', scale = 1)
plt.quiver([0], [0], *zip(*t), angles = 'xy', scale_units = 'xy', scale = 1, color = "blue")
plt.quiver([0], [0], *zip(*r2), angles = 'xy', scale_units = 'xy', scale = 1, color = "red")
plt.fill([-20, -20, 20, 20], [0, 1, 1, 0], color="blue", alpha=0.3)
plt.xlim(-20, 20)
plt.ylim(-2, 2)
plt.grid()
```

### 2.2 Column Space

Let $A \in \mathbb{R}^{m\times n}$, $x\in \mathbb{R}^n$, and $b\in \mathbb{R}^m$. The **column space** of $A$ equals the set: 

$$\mathcal{C}(A) = \{ Ax \mid x \in \mathbb{R}^n \}$$

The linearly independent columns of $A$ form the **basis**. This will be shown in the next chapter.

**Proof** The column space is a subspace:

- $0 \in S$: pick $x = 0$
- If $v, w \in S$ then $(v + w) \in S$: pick $v, w \in S$, then for some $x, y \in \mathbb{R}^n$, $v = Ax$ and $w = Ay$. But then $v + w = Ax + Ay = A(x + y)$, which is also in $S$
- If $\alpha \in \mathbb{R}$ and $v \in S$ then $\alpha v \in S$: pick $\alpha \in \mathbb{R}$ and $v \in S$, then for some $x \in \mathbb{R}^n, v = Ax$. But then $\alpha v = \alpha(Ax) = A(\alpha x)$, which is also in $S$ since $\alpha x \in \mathbb{R}^n$

What we just showed is that $S$ is a subspace, and that it is defined as a matrix multiplication. Now, here we come back at what was said earlier, in the case of no solution for $Ax = b$. It has a solution if and only if $b$ is in the **column space** of $A$. 

So, $Ax = b$ has a solution $\Longleftrightarrow b \in \mathcal{C}(A)$.

> In linear algebra, the **column space** (also called the **range** or **image**) of a matrix $A$ is the **span** (set of all possible linear combinations) of its column vectors. The column space of an m × n matrix with components from is a **linear subspace** of the m-space .

### 2.3 Null Space

Let $A \in \mathbb{R}^{m\times n}$. The set of all vectors $x \in \mathbb{R}^n$ for which $Ax = 0$ is called the **null space** of $A$:

$$\mathcal{N}(A) = \{x \mid Ax = 0\}$$

To find the null space of a matrix, we reduce it to it's echelon form. Then take combinations of the columns of $A$ for each free variable. 

$A = 
\begin{bmatrix}
1 & 2 & 2 & 2 \\
2 & 4 & 6 & 8 \\
3 & 6 & 8 & 10
\end{bmatrix}
$
$\rightarrow 
\begin{bmatrix}
1 & 2 & 2 & 2 \\
0 & 0 & 2 & 4 \\
0 & 0 & 2 & 4
\end{bmatrix}
$
$\rightarrow 
\begin{bmatrix}
1 & 2 & 2 & 2 \\
0 & 0 & 2 & 4 \\
0 & 0 & 0 & 0
\end{bmatrix}
$.

$A_1$, and $A_3$ are free columns. So $\chi_1, \chi_3$ are free variables. What this means is that we can choose them to be any value we like, and the solution to $Ax=0$ will be in the null space. So if we take a solution where we pick a free variable to be 1, and the other to be 0, we end up with a vector that is in the null space. If we make the other variable one, we have another vector. The combination of these scaled vectors construct all of the null space.

A solution would be $
\begin{pmatrix}
-2 \\
\chi_1 = 1 \\
0 \\
\chi_3 = 0
\end{pmatrix}
$ and another $
\begin{pmatrix}
2 \\
\chi_1 = 0 \\
-2 \\
\chi_3 = 1
\end{pmatrix}
$. 

So all of the null space is:
$\alpha
\begin{pmatrix}
-2 \\
1 \\
0 \\
0
\end{pmatrix}
$
$ + \beta
\begin{pmatrix}
2 \\
0 \\
-2 \\
1
\end{pmatrix}
$.

**Proof** The null space of $A$, is a subspace:

- $0 \in \mathcal{N}(A)$: pick $x = 0$
- If $x, y \in \mathcal{N}(A)$ then $x + y \in \mathcal{N}(A)$: Let $x, y \in \mathcal{N}(A)$ so that $Ax=0$, then $A(x + y) = Ax + Ay = 0 + 0 = 0$
- If $\alpha \in \mathbb{R}$ and $x \in \mathcal{N}(A)$ then $\alpha x \in \mathcal{N}(A)$: Let $\alpha \in \mathbb{R}$ and $x \in \mathcal{N}(A)$ so that $Ax = 0$, then $A(\alpha x) = A\alpha x = \alpha Ax = \alpha 0 = 0$

## 3. Span, Linear Dependence, Basis, Dimension, and Rank

Let's take a look at matrix $A$ in order to explain these concepts:

Let $A=
\begin{bmatrix}
1 & 0 & -1 & 0 & 4 \\
2 & 1 & 0 & 0 & 9 \\
-1 & 2 & 5 & 1 & -5 \\
1 & -1 & -3 & -2 & 9 \\
\end{bmatrix}
$

The column space of $A$, is equal to the **span** of its column vectors. So:

$$\mathcal{C}(A) = span(a_0, a_1, a_2, a_3)$$

The basis of $A$ is the subset of **linearly independent** in the column space of $A$. To figure out what the basis is, you put $A$ in reduced row echelon form:

$\rightarrow
\begin{bmatrix}
1 & 0 & -1 & 0 & 4 \\
0 & 1 & 2 & 0 & 1 \\
0 & 2 & 4 & 1 & -1 \\
0 & -1 & -2 & -2 & 5 \\
\end{bmatrix}
\rightarrow$
$
\begin{bmatrix}
1 & 0 & -1 & 0 & 4 \\
0 & 1 & 2 & 0 & 1 \\
0 & 0 & 0 & 1 & -3 \\
0 & 0 & 0 & -2 & 6 \\
\end{bmatrix}
\rightarrow$
$
\begin{bmatrix}
1 & 0 & -1 & 0 & 4 \\
0 & 1 & 2 & 0 & 1 \\
0 & 0 & 0 & 1 & -3 \\
0 & 0 & 0 & 0 & 0 \\
\end{bmatrix}
$

Now, $a_0, a_1,$ and $a_3$ are pivot columns. So the other columns, $a_2, a_4$, can be represented with combinations of the pivot columns. It is said that these  pivot columns are **linearly independent**.

If you take the original columns in $A$ that turned out to be the pivot columns, you end up with the **basis** for the column space of $A$.

The **dimension** of the column space of $A$, is the number of vectors in the **basis** of the column space. Which is three in this case. This even has a special name, which is the **rank** of $A$.

### 3.1 Span

The [span of a set of vectors](https://www.youtube.com/watch?v=k7RM-ot2NWY&feature=youtu.be&t=241) $V$ is the set of vectors that can be created by adding and scaling the vectors in $V$.

Given a set of vectors $\{\upsilon_0, \upsilon_1, \ldots, \upsilon_{n-1}\} \subset \mathbb{R}^n$, we can create:

$$V =
\left[
\begin{array}{c|c}
\upsilon_0 & \upsilon_1 & \cdots & \upsilon_{n-1}
\end{array}
\right]
$$

Given a matrix $V \in \mathbb{R}^{m\times n}$ and vector $x \in \mathbb{R}^n$, we have the multiplication:

$$Vx = \chi_0\upsilon_0 + \chi_1\upsilon_1 + \cdots + \chi_{n-1}\upsilon_{n-1}$$

Which is the column space of $V, \mathcal{C}(V)$:

$$
\begin{array}{ll}
\mathcal{C}(V) & = & \{Vx \mid x \in \mathbb{R}^n\} \\
& = & \{\chi_0\upsilon_0 + \chi_1\upsilon_1 + \cdots + \chi_{n-1}\upsilon_{n-1} \mid \chi_0, \chi_1, \ldots, \chi_{n-1}\in\mathbb{R}\}
\end{array}
$$

### 3.2 Linear Independence

Now if you had two vectors, you could reach every single point in 2d space. Unless these vectors were **linearly dependent**:

```python
v = [-2, 3]
w = [4, -6]

plt.quiver([0], [0], *zip(*[v, w]), angles = 'xy', scale_units = 'xy', scale = 1, color=['r', 'b'])
plt.xlim(-10, 10)
plt.ylim(-7, 7)
plt.grid()
```

Both individual vectors have the the same span of these vectors combined.

A set of vectors is **linearly dependent** if you can get the zero vector by adding or scaling individual  vectors where each scalar is not equal to zero. For example:

$$0 =$$
$$
\begin{pmatrix}
2v + w
\end{pmatrix}
=$$
$$
\begin{pmatrix}
2
\begin{pmatrix}
-2 \\
3
\end{pmatrix}
+
\begin{pmatrix}
4 \\
-6
\end{pmatrix}
\end{pmatrix}
=$$
$$
\begin{pmatrix}
0 \\
0
\end{pmatrix}
$$

### 3.3 Basis

The basis is a linearly independen spanning set.

Let $S$ be a subspace of $\mathbb{R}^m$, then the set:

$$\{\upsilon_0 + \upsilon_1 + \ldots + \upsilon_{n-1}\} \in \mathbb{R}^m$$

Is said to be a basis of $S$ if:

- $\{\upsilon_0 + \upsilon_1 + \ldots + \upsilon_{n-1}\}$ are linearly independent; and
- Span($\{\upsilon_0 + \upsilon_1 + \ldots + \upsilon_{n-1}\}) = S$;

### 3.4 Dimension

The number of vectors in the basis is called the dimension, $dim(S)$, of the subspace.

The basis of a subspace has a number of vectors in it. Multiplying all vectors by a nonzero constant results in a new basis. Therefore the basis of a subspace doesn't have to be unique, and the dimension will stay the same.

### 3.5 Rank

> The rank of a transformation is the number of dimensions in the output of a transformation.
> -- <cite>[3Blue1Brown](https://www.youtube.com/watch?v=uQhTuRlWMxw&feature=youtu.be&t=482)</cite>

Let $A \in \mathbb{R}^{m\times n}$. The rank of $A$, equals the number of vectors in a basis for the column space of $A$.

Let's look at an example to show the relation between the column space, null space, dimension and rank of the basis of a subspace:

Let $A =
\begin{pmatrix}
1 & 2 & 5 \\
0 & 3 & 6 \\
1 & 1 & 3
\end{pmatrix}
$
$ \rightarrow
\begin{pmatrix}
1 & 0 & 1 \\
0 & 1 & 2 \\
0 & 0 & 0
\end{pmatrix}
\rightarrow $
$
\begin{pmatrix}
1 & 0 & 1 \\
0 & 1 & 2 \\
0 & 0 & 0
\end{pmatrix}
$
$
\begin{pmatrix}
\chi_0 \\
\chi_1 \\
\chi_2 \\
\end{pmatrix}
=$
$
\chi_2
\begin{pmatrix}
-1 \\
-2 \\
1
\end{pmatrix}
$

- The dimension of $A$ equals three: $dim(A) = 3$
- The rank, or dimension of the column space equals two: $rank(A) = dim(\mathcal{C}(A)) = 2$
- The dimension of the null space equals one: $dim(\mathcal{N}(A)) = 1$
